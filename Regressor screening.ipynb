## Screeening regressors for wine dataset (from Kaggle)

Dataset: https://www.kaggle.com/uciml/red-wine-quality-cortez-et-al-2009

### Importing libraries

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time 
from sklearn.model_selection import train_test_split
```

```python
from lazypredict.Supervised import LazyRegressor
```

### Read the csv file

```python
dfwine = pd.read_csv ('winequality-red.csv')
```

### Data preprocessing

```python
dfwine
```

```python
#checking data integrity
dfwine.isna().count()
```

```python
data= dfwine
X=dfwine.drop('quality', axis=1)
Y=dfwine['quality']
```

```python
#Splitting data for training and testing. 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=45)
```

### Comparing Regressors

```python
clf = LazyRegressor(verbose=0,ignore_warnings=False,  custom_metric=None)
models_train,predictions_train = clf.fit(X_train, X_train, Y_train, Y_train)
```

```python
clf = LazyRegressor(verbose=0,ignore_warnings=False,  custom_metric=None)
models_test,predictions_test = clf.fit(X_train, X_test, Y_train, Y_test)

```

```python
# Performance table of the training set 
predictions_train
```

```python
# Performance table of the test set 
predictions_test
```

### Random Forest Regressor
with adjusted number of trees

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
clf = RandomForestRegressor(n_estimators=1500)
clf.fit(X_train,Y_train)
```

```python
y_pred = clf.predict(X_test)
```

*RMSE*

```python
mean_squared_error(Y_test, y_pred)
```

*R- squeared*

```python
from sklearn.metrics import r2_score
r2_score(Y_test, y_pred)
```
